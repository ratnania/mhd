# 1. GEMPIC

## 1.1 Lie-Trotter splitting without Control Variate

```python
data = np.loadtxt('/home/florian/Schreibtisch/Masterarbeit/06_Github/learning-python/data_Nel=32_dt=0.0125_Np=1e5_Nt=16000_amp=1e-4_FirstLie_NoCV')

en_B  = data[:, 0]
en_E  = data[:, 1]
en_C  = data[:, 2]
en_H  = data[:, 3]
```

```python
f1 = plt.figure()
f1.set_figheight(7)
f1.set_figwidth(10)

fontsize = 24
#plt.rc('text', usetex = False)
#plt.rc('font',family = 'serif')
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)

en_tot = en_B + en_E + en_C + en_H

Nt = len(en_B) - 1
dt = 0.05/4
tn = np.linspace(0, Nt*dt, Nt + 1) 
tg = np.linspace(0, 130, 100)

omega = utils.solveDispersionHybrid(2., +1, 1, -1, 2., 0.2, 0.53, 6e-2, 0.5 + 0j, 1e-8)[0]
omega_r = np.real(omega)
gamma = np.imag(omega)

plt.semilogy(tn, en_B/en_tot[0], linewidth = 2, label = r'$\mathcal{E}_\tilde{B}$')
plt.semilogy(tn, en_E/en_tot[0], linewidth = 2, label = r'$\mathcal{E}_\tilde{E}$')
plt.semilogy(tn, en_C/en_tot[0], '--', linewidth = 2, label = r'$\mathcal{E}_\mathrm{cold}$')
plt.semilogy(tn, en_H/en_tot[0], linewidth = 2, label = r'$\mathcal{E}_\mathrm{hot}$')
plt.semilogy(tg, 1e-6*np.exp(tg*gamma*2), 'k--', label = 'growth', linewidth = 2)

plt.xlabel('$t|\Omega_\mathrm{ce}|$')
plt.ylabel(r'$\mathcal{E} / \mathcal{E}(0)$')
plt.title('Partition of energy', fontsize = fontsize)
plt.ylim((1e-8, 1e1))
plt.xlim((0, 200))
plt.legend(loc = 4)
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/04_Ringberg/01_Figures/Energies_GEMPIC_NoCV.pdf', bbox_inches = 'tight')
```

```python
f2 = plt.figure()
f2.set_figheight(7)
f2.set_figwidth(10)

fontsize = 24
#plt.rc('text', usetex = False)
#plt.rc('font',family = 'serif')
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)

en_tot = en_B + en_E + en_C + en_H

Nt = len(en_B) - 1
dt = 0.05/4
tn = np.linspace(0, Nt*dt, Nt + 1)

plt.semilogy(tn, np.abs(en_tot - en_tot[0])/(en_tot[0]), linewidth = 2)
plt.xlim((0, 200))
plt.ylim((1e-10, 1e-2))
plt.xlabel('$t|\Omega_\mathrm{ce}|$')
plt.ylabel('$|\mathcal{E}(t) - \mathcal{E}(0)|/\mathcal{E}(0)$')
plt.title('Relative error in total energy', fontsize = fontsize)
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/04_Ringberg/01_Figures/Conservation_GEMPIC_NoCV.pdf', bbox_inches = 'tight')
```
