# Dispersion relations for cold fluid/hybrid electrons

## 1. Cold fluid electrons

Only frequencies much larger than the ion plasma frequency $\Omega_\text{pi}$ are considered why ions are treated as a neutralizing stationary background. The remaining linear model for the perturbed electric field $\textbf{E}_1$, magnetic field $\textbf{B}_1$ and electron current density $\textbf{j}_1=qn_0\textbf{u}_1$ then reads (omit subscripts):

\begin{align}
\frac{\partial \textbf{j}}{\partial t} - \epsilon_0\Omega_\text{pe}^2\textbf{E} + \Omega_0\times\textbf{j} = 0, \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} - \nabla\times\textbf{B} + \mu_0\textbf{j} = 0, \\
\frac{\partial \textbf{B}}{\partial t} + \nabla\times\textbf{E} = 0,
\end{align}

with $\Omega_0=\frac{q\textbf{B}_0}{m}$ and $\Omega_\text{pe}^2=\frac{q^2 n_0}{\epsilon_0 m}$.

Assuming that all quantites ~$e^{i(kz-\omega t)}$ and $\textbf{B}_0=B_0\textbf{e}_z$ the following dispersion relations can be obtained for R- and L-waves, respectivley:

\begin{align}
D_\text{cold,R/L}(k,\omega)=1-\frac{c^2k^2}{\omega^2}-\frac{\Omega_\text{pe}^2}{\omega(\omega\pm\Omega_\text{ce})}=0 \Leftrightarrow k_\text{R/L} = \frac{\omega}{c}\sqrt{1-\frac{\Omega_\text{pe}^2}{\omega(\omega\pm\Omega_\text{ce})}}, 
\end{align}

where $\Omega_\text{ce}=\frac{qB_0}{m}=-\frac{eB_0}{m}<0$ for electrons.




## 2. Cold fluid electrons with artificial energetic particles

In order to get a damping/growth of the wave field, it is assumed that there is current density $\textbf{j}_\text{h}$ coming from energetic particles. For simplicity it is assumed that the there is linear response to the electric field of the from

\begin{align}
\textbf{j}_\text{h} = \sigma_\text{h}\textbf{E},
\end{align}

with some constant conductivity tensor

\begin{align}
\sigma_\text{h}=
\begin{pmatrix}
c_1 & c_2 &0 \\
-c_2 &c_1 &0 \\
0 &0 &c_3 \\
\end{pmatrix}.
\end{align}

The model then reads

\begin{align}
\frac{\partial \textbf{j}}{\partial t} - \epsilon_0\Omega_\text{pe}^2\textbf{E} + \Omega_0\times\textbf{j} = 0, \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} - \nabla\times\textbf{B} + \mu_0(\textbf{j}+\sigma_\text{h}\textbf{E}) = 0, \\
\frac{\partial \textbf{B}}{\partial t} + \nabla\times\textbf{E} = 0,
\end{align}

with the dispersion relation

\begin{align}
D_\text{art,R/L}(k,\omega)=D_\text{cold,R/L}(k,\omega)+\frac{c^2\mu_0}{\omega}(ic_1\mp c_2)
\end{align}

for R/L-waves ($c_3=0$).



## 3. Hybrid model
The self-consistend electron hybrid model with energetic electrons which are treated kinetically reads

\begin{align}
\frac{\partial \textbf{j}_\text{c}}{\partial t} = \epsilon_0\Omega_{\text{pe}}^2\textbf{E} + \Omega_\text{ce}\textbf{j}_\text{c}\times\textbf{e}_z,  \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} = \nabla\times\textbf{B} - \mu_0(\textbf{j}_\text{c} + \textbf{j}_\text{h}), \\
\frac{\partial \textbf{B}}{\partial t}=-\nabla\times\textbf{E}, \\
\frac{\partial f_\text{h}}{\partial t} + \textbf{v}\cdot\nabla f_\text{h} + \frac{q}{m}(\textbf{E}+\textbf{v}\times\textbf{B})\cdot\nabla_v f_\text{h}=0, \\
\textbf{j}_\text{h} = q\int\text{d}^3\textbf{v}\,\textbf{v}f_\text{h}.
\end{align}

Linear theory of the above model leads to the following general dispersion relation for an arbitrary equilibrium distribution function $f^0=f^0(v_\parallel,v_\bot)$:


\begin{align}
D_{\text{R/L}}(k,\omega)=1-\frac{c^2k^2}{\omega^2}-\frac{\Omega_{\text{pe}}^2}{\omega(\omega\pm\Omega_{\text{ce}})}+\nu_\text{h}\frac{\Omega_{\text{pe}}^2}{\omega}\int\text{d}^3\textbf{v}\frac{v_\bot}{2}\frac{\hat{G}f_\text{h}^0}{\omega\pm\Omega_{\text{ce}}-kv_\parallel}=0.
\end{align}

Here $\nu_\text{h}=n_\text{h}/n_\text{c}\ll1$ is the ratio between the hot and cold electron number densities, respectively, $\text{d}^3\textbf{v}=\text{d}v_\parallel\text{d}v_\bot v_\bot 2\pi$ and the differential operator

\begin{align}
\hat{G}=\frac{\partial}{\partial v_\bot}+\frac{k}{\omega}\left(v_\bot\frac{\partial}{\partial v_\parallel}-v_\parallel\frac{\partial}{\partial v_\bot}\right).
\end{align}

For an anisotropic Maxwellian 

\begin{align}
f^0(v_\parallel,v_\bot) = \frac{1}{(2\pi)^{3/2}w_\parallel w_\bot^2}\exp\left(-\frac{v_\parallel^2}{2w_\parallel^2}-\frac{v_\bot^2}{2w_\bot^2}\right)
\end{align}

the dispersion relation is given by

\begin{align}
D_{\text{R/L}}(k,\omega)=D_{\text{cold,R/L}}(k,\omega)+\nu_\text{h}\frac{\Omega_{\text{pe}}^2}{\omega^2}\left[\frac{\omega}{k\sqrt{2}w_\parallel}Z(\xi^{\pm})-\left(1-\frac{w_\bot^2}{w_\parallel^2}\right)(1+\xi^{\pm} Z(\xi^{\pm}))\right]=0, 
\end{align}

where $Z$ is the plasma dispersion function and 

\begin{align}
\xi^{\pm} = \frac{\omega\pm\Omega_\text{ce}}{k\sqrt{2}w_\parallel}.
\end{align}

# 1. Cold plasma

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
import Utilitis_HybridCode as utils

# ... parameters
wce = -1                           # ... electron cyclotron frequency
wpe = 5*np.abs(wce)                # ... cold plasma frequency (normalized to electron cyclotron frequency) 
wpar = 0.2                         # ... parallel thermal velocity of hot particles (normalized to speed of light)
wperp = 0.53                       # ... perpendiculer thermal velocity of hot particles (normalized to speed of light)
nuh = 6e-3                         # ... density ratio of cold/hot particles
c1 = -0.5                          # ... artificial energetic current 
c2 = -0.1                          # ... artificial energetic current 
mu0 = 1                            # ... vacuum permeability
c = 1                              # ... speed of light
# ...



# ... plasma dispersion function and its derivative
def Z(xi):
    return np.sqrt(np.pi)*np.exp(-xi**2)*(1j - sp.erfi(xi))

def Zprime(xi):
    return -2*(1 + xi*Z(xi))
# ...




# ... Newton method for cold plasma dispersion relation for fixed k
k1 = np.linspace(0.1,10.0,40)

w1_1 = np.zeros(40)
w1_2 = np.zeros(40)
w1_3 = np.zeros(40)

w1_1[0] = 0.0001 
w1_2[0] = 2.0001 
w1_3[0] = 2.5001


for i in range(0,40):
    
    if i == 0:
        w1_1[i] = utils.solveDispersionCold(k1[i],+1,c,wce,wpe,w1_1[i],1e-6,100)[0]
        w1_2[i] = utils.solveDispersionCold(k1[i],-1,c,wce,wpe,w1_2[i],1e-6,100)[0]
        w1_3[i] = utils.solveDispersionCold(k1[i],+1,c,wce,wpe,w1_3[i],1e-6,100)[0]
    
    else:
        w1_1[i] = utils.solveDispersionCold(k1[i],+1,c,wce,wpe,w1_1[i-1],1e-6,100)[0]
        w1_2[i] = utils.solveDispersionCold(k1[i],-1,c,wce,wpe,w1_2[i-1],1e-6,100)[0]
        w1_3[i] = utils.solveDispersionCold(k1[i],+1,c,wce,wpe,w1_3[i-1],1e-6,100)[0]
    
# ...


# ... plot results
f1 = plt.figure()
f1.set_figheight(8)
f1.set_figwidth(12)
plt.subplots_adjust(wspace = 0.3, hspace = 0.4)


plt.rc('text',usetex = True)
plt.rc('font',family = 'serif')
plt.rcParams.update({'font.size': 16})
plt.rc('xtick',labelsize = 16)
plt.rc('ytick',labelsize = 16)

plt.subplot(221)
plt.plot(k1,np.real(w1_1),label = 'R-wave')
plt.plot(k1,np.real(w1_2),label = 'L-wave')
plt.plot(k1,np.real(w1_3),label = 'R-wave')
plt.ylim((0,4))
plt.plot(k1,np.ones(len(k1))*(-wce),'k--',label = r'$\Omega_\mathrm{ce}$')
plt.title('Cold plasma dispersion relation',fontsize = 16)
plt.xlabel(r'$kc / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Re}(\omega) / \Omega_\mathrm{ce}$')
plt.legend()
# ... 
```

```python
# ... Newton method for artificial plasma dispersion relation for fixed k
k2 = np.linspace(0.1,8.0,40)

w2_1 = np.zeros(40,dtype = complex)
w2_2 = np.zeros(40,dtype = complex)
w2_3 = np.zeros(40,dtype = complex)

w2_1[0] = 0.0001 + 0j
w2_2[0] = 2.0001 + 0j
w2_3[0] = 2.5001 + 0j


for i in range(0,40):
    
    if i == 0:
        w2_1[i] = utils.solveDispersionArtificial(k2[i],+1,c,wce,wpe,c1,c2,mu0,w2_1[i],1e-6,100)[0]
        w2_2[i] = utils.solveDispersionArtificial(k2[i],-1,c,wce,wpe,c1,c2,mu0,w2_2[i],1e-6,100)[0]
        w2_3[i] = utils.solveDispersionArtificial(k2[i],+1,c,wce,wpe,c1,c2,mu0,w2_3[i],1e-6,100)[0]
    
    else:
        w2_1[i] = utils.solveDispersionArtificial(k2[i],+1,c,wce,wpe,c1,c2,mu0,w2_1[i-1],1e-6,100)[0]
        w2_2[i] = utils.solveDispersionArtificial(k2[i],-1,c,wce,wpe,c1,c2,mu0,w2_2[i-1],1e-6,100)[0]
        w2_3[i] = utils.solveDispersionArtificial(k2[i],+1,c,wce,wpe,c1,c2,mu0,w2_3[i-1],1e-6,100)[0]
    
# ...


# ... plot results
f2 = plt.figure()
f2.set_figheight(8)
f2.set_figwidth(12)
plt.subplots_adjust(wspace = 0.3, hspace = 0.4)


plt.subplot(221)
plt.plot(k2,np.real(w2_1),label = 'R-wave')
plt.plot(k2,np.real(w2_2),label = 'L-wave')
plt.plot(k2,np.real(w2_3),label = 'R-wave')
plt.plot(k1,np.real(w1_1),'--',label = 'cold plasma')
plt.ylim((0,4))
plt.plot(k2,np.ones(len(k2))*(-wce),'k--')
plt.xlabel(r'$kc / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Re}(\omega) / \Omega_\mathrm{ce}$')
plt.title('Real oscillation frequency',fontsize = 16)
plt.legend()


plt.subplot(222)
plt.plot(k2,np.imag(w2_1),label = 'R-wave: artificial')
plt.plot(k2,np.imag(w2_2),label = 'L-wave: artificial')
plt.plot(k2,np.imag(w2_3),label = 'R-wave: artificial')
plt.xlabel(r'$kc / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Im}(\omega) / \Omega_\mathrm{ce}$')
plt.title('Growth rate',fontsize = 16)
plt.legend()


plt.subplot(223)
plt.plot(np.real(w2_1),np.imag(w2_1))
plt.plot(np.real(w2_2),np.imag(w2_2))
plt.plot(np.real(w2_3),np.imag(w2_3))
plt.title('Growth rate',fontsize = 16)
plt.xlabel(r'$\mathrm{Re}(\omega) / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Im}(\omega) / \Omega_\mathrm{ce}$')
# ...
```

# 3. Hybrid model

```python
# ... Newton method for hybrid plasma dispersion relation for fixed k
k3 = np.linspace(0.4,8.0,100)

w3_1 = np.zeros(len(k3),dtype = complex)
w3_2 = np.zeros(len(k3),dtype = complex)
w3_3 = np.zeros(len(k3),dtype = complex)

w3_1[0] = 0.0001 + 0j
w3_2[0] = 2.0001 + 0j
w3_3[0] = 2.5001 + 0j

gamma_explicit = np.zeros(len(k3))


for i in range(0,len(k3)):
    
    if i == 0:
        w3_1[i] = utils.solveDispersionHybrid(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,w3_1[i],1e-6,100)[0]
        w3_2[i] = utils.solveDispersionHybrid(k3[i],-1,c,wce,wpe,wpar,wperp,nuh,w3_2[i],1e-6,100)[0]
        w3_3[i] = utils.solveDispersionHybrid(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,w3_3[i],1e-6,100)[0]
        
        gamma_explicit[i] = utils.solveDispersionHybridExplicit(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,np.real(w3_1[i]),1e-6,1000)[1]
    
    else:
        w3_1[i] = utils.solveDispersionHybrid(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,w3_1[i-1],1e-6,100)[0]
        w3_2[i] = utils.solveDispersionHybrid(k3[i],-1,c,wce,wpe,wpar,wperp,nuh,w3_2[i-1],1e-6,100)[0]
        w3_3[i] = utils.solveDispersionHybrid(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,w3_3[i-1],1e-6,100)[0]
    
        gamma_explicit[i] = utils.solveDispersionHybridExplicit(k3[i],+1,c,wce,wpe,wpar,wperp,nuh,np.real(w3_1[i]),1e-6,1000)[1]
# ...



# ... plot results
f2 = plt.figure()
f2.set_figheight(8)
f2.set_figwidth(12)
plt.subplots_adjust(wspace = 0.3, hspace = 0.4)


def plot_DispersionFunction():
    plt.subplot(221)
    xi = np.linspace(-4,4,100)
    plt.plot(xi,np.real(Z(xi)),label = r'Re($Z(\xi)$)')
    plt.plot(xi,np.imag(Z(xi)),label = r'Im($Z(\xi)$)')
    plt.title(r'Plasma dispersion function $Z$',fontsize = 16)
    plt.xlabel(r'$\xi$')
    plt.ylim((-2,2))
    plt.legend()



    plt.subplot(222)
    plt.plot(xi,np.real(Zprime(xi)),label = r'Re($Z^\prime(\xi)$)')
    plt.plot(xi,np.imag(Zprime(xi)),label = r'Im($Z^\prime(\xi)$)')
    plt.title(r'Derivative of Plasma dispersion function $\frac{\mathrm{d}Z}{\mathrm{d}\xi}$',fontsize = 16)
    plt.xlabel(r'$\xi$')
    plt.ylim((-2.5,2.5))
    plt.legend()


plot_DispersionFunction()


plt.subplot(223)
plt.plot(k3,np.real(w3_1),label = 'R-wave',linewidth = 2)
plt.plot(k3,np.real(w3_2),label = 'L-wave',linewidth = 2)
plt.plot(k3,np.real(w3_3),label = 'R-wave',linewidth = 2)
plt.plot(k1,np.real(w1_1),'r--',label = 'cold plasma')
plt.ylim((0,4))
plt.plot(k1,np.ones(len(k1))*(-wce),'k--')
plt.xlabel(r'$kc / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Re}(\omega) / \Omega_\mathrm{ce}$')
plt.title('Real oscillation frequency',fontsize = 16)
plt.legend()



plt.subplot(224)
plt.plot(k3,np.imag(w3_1),label = 'R-wave',linewidth = 2)
plt.plot(k3,np.imag(w3_2),label = 'L-wave',linewidth = 2)
plt.plot(k3,np.imag(w3_3),label = 'R-wave',linewidth = 2)
plt.plot(k3,gamma_explicit,'k--',label = 'R-wave - explicit',linewidth = 2)
plt.xlabel(r'$kc / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Im}(\omega) / \Omega_\mathrm{ce}$')
plt.title('Growth rate',fontsize = 16)
plt.legend()
# ...


# ... save figure as .pdf
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/02_Summary_HalfTime/01_Figures/HybridDispersion.pdf',bbox_inches = 'tight')
# ...
```

# 3.1 Maximum growth rate for different anisotropies

```python
wperp = np.linspace(0.3, 0.7, 10)

k4 = np.linspace(0.4, 8.0, 800)
w4 = np.zeros((len(wperp), len(k4)), dtype = complex)
w4[:, 0] = 0.0001 + 0j 
max_growth = np.zeros(len(wperp))
max_freq = np.zeros(len(wperp))
ind_max = np.zeros(len(wperp), dtype = int)

for j in range(len(wperp)):
    
    for i in range(0, len(k4)):

        if i == 0:
            w4[j, i] = utils.solveDispersionHybrid(k4[i], +1, c, wce, wpe, wpar, wperp[j], nuh, w4[j, i], 1e-6, 100)[0]
            
        else:
            w4[j, i] = utils.solveDispersionHybrid(k4[i], +1, c, wce, wpe, wpar, wperp[j], nuh, w4[j, i-1], 1e-6, 100)[0]
            

    plt.plot(np.real(w4[j]), np.imag(w4[j]))
    ind_max[j] = np.where(np.imag(w4[j]) == np.max(np.imag(w4[j])))[0]
    max_growth[j] = np.imag(w4[j, ind_max[j]])
    max_freq[j] = np.real(w4[j, ind_max[j]])
    
plt.xlabel(r'$\mathrm{Re}(\omega) / \Omega_\mathrm{ce}$')
plt.ylabel(r'$\mathrm{Im}(\omega) / \Omega_\mathrm{ce}$')
```
